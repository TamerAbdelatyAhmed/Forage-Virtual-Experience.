## Loading Datasets into DataFrames:

```python
content = pd.read_csv("Content.csv",  index_col = [0])
reactions = pd.read_csv("Reactions.csv", index_col = [0])
reaction_types = pd.read_csv("ReactionTypes.csv", index_col = [0])
```

```python
# remove the quotation marks from the 'Category' column
content['Category'] = content['Category'].str.replace('"', '')
content['Category']= content['Category'].str.lower()
content.drop(columns="URL", inplace=True)
# print the updated dataframe
content
```

```python
# remane column "Type" name as "content_type" in the three dataframes:
def rename_col(dataframe, col_name, new_col_name):
    dataframe.rename(columns={col_name: new_col_name}, inplace=True)
    
rename_content_type_col = rename_col(content, "Type", "content_type")
rename_reactions_type_col = rename_col(reactions, "Type", "reaction_type")    
rename_reaction_types_type_col = rename_col(reaction_types, "Type", "reaction_type")  
```

```python
content.drop(columns="User ID", inplace=True)
reactions.drop(columns="User ID", inplace=True)
```

```python
# merge the dataframes
data = pd.merge(reactions, reaction_types, on='reaction_type', how='left')
data = pd.merge(data, content, on=['Content ID'], how='right')
data.dropna(inplace=True)
# print the resulting dataframe
data.head()
```

```python
data.info()
```

```python
total_categories = data.groupby(by = 'Category', as_index= False)['Score'].sum()
total_categories_sort = total_categories.sort_values("Score", ascending=False)
total_categories_sort
```

```python
category_scores = data.groupby('Category')['Score'].sum().reset_index()

top_categories = category_scores.sort_values('Score', ascending=False).head(5)

top_categories
```

```python
data.groupby('Category').agg({"reaction_type" : "count"}).reset_index().sort_values(by='reaction_type', ascending=False).head(5)
```

```python
top_categories.to_csv("The top 5 categories.csv")
```

```python
# get the top 5 categories by frequency
top_categories = data.groupby("Category")['Score'].value_counts().nlargest(5)

# print the top 5 categories
print(top_categories)
```

```python
data.to_csv("cleaned_merged_data.csv")
```

```python
# get the top 5 categories by frequency
top_categories = data.groupby('Category')['Score'].value_counts().nlargest(5)

# create a mask of rows with the top categories and scores
mask = data.set_index(['Category', 'Score']).index.isin(top_categories.index)

# apply the mask to the data and create a new dataframe
data1 = data[mask]

# print the resulting dataframe
data1.info()
data1.head()
```

```python
data1.to_csv("The top 5 categories_all.csv")
```
